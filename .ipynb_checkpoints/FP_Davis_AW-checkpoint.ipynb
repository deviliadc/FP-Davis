# DISTRIBUTION

```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt

# Define the database connection string using SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# Query to retrieve sales data over time
query = """
SELECT 
    t.CalendarYear AS years, 
    COUNT(sf.OrderQuantity) AS order_count
FROM  
    dw_aw.factinternetsales sf
JOIN 
    dw_aw.dimtime t ON sf.OrderDateKey = t.TimeKey
GROUP BY 
    t.CalendarYear
ORDER BY 
    t.CalendarYear;
"""

try:
    # Read data from the database
    data = pd.read_sql(query, db_connection)
    print(data)

    # Plot the data
    plt.plot(data['years'], data['order_count'], marker='o')

    plt.title("Trend of Order Count Over Years")

    plt.xlabel('Year')
    plt.ylabel('Order Count')

    plt.legend(['Order Count'])

    plt.show()

except Exception as e:
    print(str(e))
```

# COMPOSITION

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify
from sqlalchemy import create_engine

# Define the database connection string using SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# SQL query to retrieve the number of customers per country
query = """
SELECT 
    g.EnglishCountryRegionName AS Country,
    COUNT(c.CustomerKey) AS CustomerCount
FROM 
    dimcustomer c
JOIN 
    dimgeography g ON c.GeographyKey = g.GeographyKey
GROUP BY 
    Country
ORDER BY 
    CustomerCount DESC;
"""

# Read data from the database using pandas
df = pd.read_sql(query, db_connection)

# Sort the dataframe based on the number of customers in descending order
df = df.sort_values(by='CustomerCount', ascending=False)

# Calculate the area size for each box in the treemap
sizes = df['CustomerCount'].values.tolist()

# Prepare labels for each box
labels = ["{}\n({} customers)".format(country, count) for country, count in zip(df['Country'], df['CustomerCount'])]

# Create the treemap
plt.figure(figsize=(10, 6))
squarify.plot(sizes=sizes, label=labels, alpha=0.7)

# Add title and remove axes
plt.title("Distribution of Customers by Country")
plt.axis('off')

# Display the treemap
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine

# Connection string to the database using SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# SQL query to retrieve ListPrice, StandardCost, and count of records
query = """
SELECT 
    ListPrice,
    StandardCost,
    COUNT(*) AS Count
FROM 
    dimproduct
GROUP BY 
    ListPrice,
    StandardCost;
"""

# Read data from the database using pandas
df = pd.read_sql(query, db_connection)

# Create scatter plot with margins
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='StandardCost', y='ListPrice', size='Count', sizes=(20, 200), alpha=0.7)

# Add title and axis labels
plt.title('Scatter Plot of Standard Cost vs List Price with Margins')
plt.xlabel('Standard Cost')
plt.ylabel('List Price')

# Show plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Connection string to the database using SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# SQL query to retrieve ListPrice and StandardCost
query = """
SELECT 
    ListPrice,
    StandardCost
FROM 
    dimproduct;
"""

# Read data from the database using pandas
df = pd.read_sql(query, db_connection)

# Calculate the margin
df['Margin'] = df['ListPrice'] - df['StandardCost']

# Calculate the average margin
average_margin = df['Margin'].mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['ListPrice'], df['Margin'], color='blue', linestyle='-', marker='o', label='Margin')

# Add horizontal line for the average margin
plt.axhline(y=average_margin, color='red', linestyle='--', label=f'Average Margin: {average_margin:.2f}')

# Add title and axis labels
plt.title('Margin between ListPrice and StandardCost')
plt.xlabel('ListPrice')
plt.ylabel('Margin')

# Add legend
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Membuat koneksi ke database menggunakan SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# Query SQL untuk mengambil data penjualan produk
query = """
SELECT 
    psc.EnglishProductSubcategoryName,
    fs.SalesAmount,
    pc.EnglishProductCategoryName
FROM 
    factinternetsales fs
JOIN 
    dimproduct dp ON fs.ProductKey = dp.ProductKey
JOIN 
    dimproductsubcategory psc ON dp.ProductSubcategoryKey = psc.ProductSubcategoryKey
JOIN 
    dimproductcategory pc ON psc.ProductCategoryKey = pc.ProductCategoryKey;
"""

# Membaca data dari database menggunakan pandas
df = pd.read_sql(query, db_connection)

# Memvisualisasikan data menggunakan bubble plot
plt.figure(figsize=(12, 8))

plt.scatter(df['EnglishProductSubcategoryName'], df['SalesAmount'], s=df['EnglishProductCategoryName'].map(len)*200, alpha=0.5)

plt.title('Bubble Plot: Penjualan Produk Subkategori dari Kategori Produk')
plt.xlabel('Nama Subkategori Produk')
plt.ylabel('Total Penjualan')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()

# Menampilkan bubble plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Membuat koneksi ke database menggunakan SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# Query SQL untuk mengambil data penjualan produk dengan variabel tambahan
query = """
SELECT 
    psc.EnglishProductSubcategoryName,
    SUM(fs.SalesAmount) AS TotalSales,
    pc.EnglishProductCategoryName
FROM 
    factinternetsales fs
JOIN 
    dimproduct dp ON fs.ProductKey = dp.ProductKey
JOIN 
    dimproductsubcategory psc ON dp.ProductSubcategoryKey = psc.ProductSubcategoryKey
JOIN 
    dimproductcategory pc ON psc.ProductCategoryKey = pc.ProductCategoryKey
GROUP BY 
    psc.EnglishProductSubcategoryName, pc.EnglishProductCategoryName;
"""

# Membaca data dari database menggunakan pandas
df = pd.read_sql(query, db_connection)

# Membuat bubble plot dengan variabel tambahan
plt.figure(figsize=(12, 8))
for category, data in df.groupby('EnglishProductCategoryName'):
    plt.scatter(data.index, data['TotalSales'], s=data['TotalSales']/1000, alpha=0.5, label=category)

plt.title('Bubble Plot Penjualan Produk Berdasarkan Subkategori dengan Variabel Kategori Produk')
plt.xlabel('Subkategori Produk')
plt.ylabel('Total Penjualan (ribuan)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Membuat koneksi ke database menggunakan SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# Query SQL untuk mengambil data penjualan berdasarkan tipe promosi dan diskon
query = """
SELECT 
    p.EnglishPromotionType AS PromotionType,
    p.DiscountPct AS Discount,
    fs.SalesAmount AS TotalSales
FROM 
    dimpromotion p
JOIN 
    factinternetsales fs ON p.PromotionKey = fs.PromotionKey;
"""

# Membaca data dari database menggunakan pandas
df = pd.read_sql(query, db_connection)

# Membuat scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['PromotionType'], df['Discount'], s=df['TotalSales'] / 1000, c=df['TotalSales'], cmap='coolwarm', alpha=0.7)

# Menambahkan judul dan label sumbu
plt.title('Distribution of Sales by Promotion Type and Discount')
plt.xlabel('Promotion Type')
plt.ylabel('Discount (%)')

# Menambahkan colorbar untuk menunjukkan total penjualan
cbar = plt.colorbar()
cbar.set_label('Total Sales')

# Menampilkan plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Membuat koneksi ke database menggunakan SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# Query SQL untuk mengambil data penjualan berdasarkan promosi
query = """
SELECT 
    dp.EnglishPromotionName AS PromotionName,
    SUM(fs.SalesAmount) AS TotalSales
FROM 
    factinternetsales fs
JOIN 
    dimpromotion dp ON fs.PromotionKey = dp.PromotionKey
GROUP BY 
    dp.EnglishPromotionName
ORDER BY 
    TotalSales DESC;
"""

# Membaca data dari database menggunakan pandas
df = pd.read_sql(query, db_connection)

# Memvisualisasikan distribusi penjualan berdasarkan promosi menggunakan bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['PromotionName'], df['TotalSales'], color='skyblue')
plt.title('Distribution of Sales by Promotion')
plt.xlabel('Promotion')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()

# Menampilkan bar chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Menentukan string koneksi database menggunakan SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# Kueri SQL untuk mengambil jumlah penjualan (SalesAmount) untuk setiap kategori produk
query = """
SELECT 
    pc.EnglishProductCategoryName AS ProductCategory,
    SUM(fs.SalesAmount) AS TotalSales
FROM 
    dimproduct p
JOIN 
    dimproductsubcategory ps ON p.ProductSubcategoryKey = ps.ProductSubcategoryKey
JOIN 
    dimproductcategory pc ON ps.ProductCategoryKey = pc.ProductCategoryKey
JOIN 
    factinternetsales fs ON p.ProductKey = fs.ProductKey
GROUP BY 
    pc.EnglishProductCategoryName
ORDER BY 
    TotalSales DESC;
"""

# Membaca data dari database menggunakan pandas
df = pd.read_sql(query, db_connection)

# Membuat diagram batang
plt.figure(figsize=(10, 6))
plt.bar(df['ProductCategory'], df['TotalSales'])

# Menambahkan judul dan label sumbu
plt.title('Comparison of Sales Amount Across Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Total Sales Amount')

# Menampilkan plot
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Menentukan string koneksi database menggunakan SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# Kueri SQL untuk mengambil top 3 penjualan produk berdasarkan wilayah
query = """
SELECT 
    st.SalesTerritoryRegion AS SalesTerritory,
    p.EnglishProductName AS ProductName,
    SUM(fs.SalesAmount) AS TotalSales
FROM 
    dimsalesterritory st
JOIN 
    factinternetsales fs ON st.SalesTerritoryKey = fs.SalesTerritoryKey
JOIN 
    dimproduct p ON fs.ProductKey = p.ProductKey
GROUP BY 
    st.SalesTerritoryRegion, p.EnglishProductName
ORDER BY 
    st.SalesTerritoryRegion, TotalSales DESC;
"""

# Membaca data dari database menggunakan pandas
df = pd.read_sql(query, db_connection)

# Membuat visualisasi untuk setiap wilayah penjualan dengan tata letak subplot 5x2
unique_territories = df['SalesTerritory'].unique()
num_territories = len(unique_territories)
num_rows = 5
num_cols = 2
num_plots = min(num_territories, num_rows * num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 20))
fig.subplots_adjust(hspace=0.5)

for i, territory in enumerate(unique_territories[:num_plots]):
    # Filter data untuk wilayah penjualan tertentu
    territory_df = df[df['SalesTerritory'] == territory]
    
    # Mengambil top 3 produk berdasarkan total penjualan
    top_3_products = territory_df.groupby('ProductName')['TotalSales'].sum().nlargest(3)
    
    # Membuat diagram batang horizontal
    ax = axes[i // num_cols, i % num_cols]
    ax.bar(top_3_products.index, top_3_products.values, color='skyblue')
    
    # Menambahkan judul dan label sumbu
    ax.set_title(f'Top 3 Sales Products in {territory}')
    ax.set_xlabel('Total Sales')
    ax.set_ylabel('Product')
    
# Menghapus subplot yang tidak digunakan
for i in range(num_plots, num_rows * num_cols):
    fig.delaxes(axes[i // num_cols, i % num_cols])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Membuat koneksi ke database menggunakan SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# Query SQL untuk mengambil data penjualan dari waktu ke waktu berdasarkan segmen pelanggan
query = """
SELECT 
    t.CalendarYear,
    c.EnglishEducation AS CustomerSegment,
    SUM(fs.SalesAmount) AS TotalSales
FROM 
    factinternetsales fs
JOIN 
    dimcustomer c ON fs.CustomerKey = c.CustomerKey
JOIN 
    dimtime t ON fs.OrderDateKey = t.TimeKey
GROUP BY 
    t.CalendarYear, c.EnglishEducation
ORDER BY 
    t.CalendarYear, TotalSales DESC;
"""

# Membaca data dari database menggunakan pandas
df = pd.read_sql(query, db_connection)

# Memvisualisasikan data menggunakan line chart
plt.figure(figsize=(12, 6))

# Menggunakan loop untuk membuat garis untuk setiap segmen pelanggan
for segment, data in df.groupby('CustomerSegment'):
    plt.plot(data['CalendarYear'], data['TotalSales'], marker='o', label=segment)

plt.title('Perbandingan Kinerja Penjualan Antar Segmen Pelanggan dari Waktu ke Waktu')
plt.xlabel('Tahun')
plt.ylabel('Total Penjualan')
plt.legend()
plt.grid(True)
plt.xticks(df['CalendarYear'].unique())  # Memberi label sumbu X dengan tahun-tahun unik
plt.tight_layout()

# Menampilkan line chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Establish connection to the database using SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# SQL query to retrieve sales data over time by region
query = """
SELECT 
    t.CalendarYear,
    g.EnglishCountryRegionName AS Region,
    SUM(fs.OrderQuantity) AS TotalOrder
FROM 
    factinternetsales fs
JOIN 
    dimcustomer c ON fs.CustomerKey = c.CustomerKey
JOIN 
    dimgeography g ON c.GeographyKey = g.GeographyKey
JOIN 
    dimtime t ON fs.OrderDateKey = t.TimeKey
GROUP BY 
    t.CalendarYear, g.EnglishCountryRegionName
ORDER BY 
    t.CalendarYear, TotalOrder DESC;
"""

# Read data from the database using pandas
df = pd.read_sql(query, db_connection)

# Visualize the data using a line chart
plt.figure(figsize=(12, 6))

# Use a loop to create lines for each region
for region, data in df.groupby('Region'):
    plt.plot(data['CalendarYear'], data['TotalOrder'], marker='o', label=region)

plt.title('Comparison of Orders Across Regions Over Time')
plt.xlabel('Year')
plt.ylabel('Total Order')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.xticks(df['CalendarYear'].unique())  # Label the X-axis with unique years
plt.tight_layout()

# Show the line chart
plt.show()
```

# Menggabungkan data dan export ke CSV

```python
import pandas as pd
from sqlalchemy import create_engine

# Membuat koneksi ke database menggunakan SQLAlchemy
db_connection_str = 'mysql+mysqlconnector://root:@localhost:3306/dw_aw'
db_connection = create_engine(db_connection_str)

# inisialisasi tabel database
factinternetsales = pd.read_sql('SELECT * FROM factinternetsales', con=db_connection)
dimtime = pd.read_sql('SELECT * FROM dimtime', con=db_connection)
dimproduct = pd.read_sql('SELECT * FROM dimproduct', con=db_connection)
dimproductsubcategory = pd.read_sql('SELECT * FROM dimproductsubcategory', con=db_connection)
dimproductcategory = pd.read_sql('SELECT * FROM dimproductcategory', con=db_connection)
dimpromotion = pd.read_sql('SELECT * FROM dimpromotion', con=db_connection)
dimcustomer = pd.read_sql('SELECT * FROM dimcustomer', con=db_connection)
dimgeography = pd.read_sql('SELECT * FROM dimgeography', con=db_connection)

# menggabungkan data tabel factinternetsales dan dimtime
merge_sales_time = pd.merge(
    left=factinternetsales,
    right=dimtime,
    how="left",
    left_on="OrderDateKey",
    right_on="TimeKey",
    suffixes=('', '_dimtime')
)

# menggabungkan dengan tabel dimproduct
merge_product = pd.merge(
    left=merge_sales_time,
    right=dimproduct,
    how="left",
    left_on="ProductKey",
    right_on="ProductKey",
    suffixes=('', '_dimproduct')
)

# menggabungkan dengan tabel dimproductsubcategory
merge_productsubcategory = pd.merge(
    left=merge_product,
    right=dimproductsubcategory,
    how="left",
    left_on="ProductSubcategoryKey",
    right_on="ProductSubcategoryKey",
    suffixes=('', '_dimproductsubcategory')
)

# menggabungkan dengan tabel dimproductcategory
merge_productcategory = pd.merge(
    left=merge_productsubcategory,
    right=dimproductcategory,
    how="left",
    left_on="ProductCategoryKey",
    right_on="ProductCategoryKey",
    suffixes=('', '_dimproductcategory')
)

# menggabungkan dengan dimpromotion
merge_promotion = pd.merge(
    left=merge_productcategory,
    right=dimpromotion,
    how="left",
    left_on="PromotionKey",
    right_on="PromotionKey",
    suffixes=('', '_dimpromotion')
)

# menggabungkan dengan customer
merge_customer = pd.merge(
    left=merge_promotion,
    right=dimcustomer,
    how="left",
    left_on="CustomerKey",
    right_on="CustomerKey",
    suffixes=('', '_dimcustomer')
)

# menggabungkan dengan dimgeography
merge_geography = pd.merge(
    left=merge_customer,
    right=dimgeography,
    how="left",
    left_on="GeographyKey",
    right_on="GeographyKey",
    suffixes=('', '_dimgeography')
)

# Memilih kolom yang unik (tidak duplikat)
unique_columns = [col for col in merge_geography.columns if not col.endswith('_dimtime') 
                  and not col.endswith('_dimproduct') 
                  and not col.endswith('_dimproductsubcategory') 
                  and not col.endswith('_dimproductcategory') 
                  and not col.endswith('_dimpromotion') 
                  and not col.endswith('_dimcustomer') 
                  and not col.endswith('_dimgeography')]

final_df = merge_geography[unique_columns]

# Menyimpan dataframe ke dalam file CSV
final_df.to_csv('aw_all.csv', index=False)

# Melihat beberapa baris dari dataframe yang sudah digabungkan
print(final_df.head())
```
